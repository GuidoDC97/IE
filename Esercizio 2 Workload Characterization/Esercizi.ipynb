```python
# Import relavant modules
import random as rnd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import pylab
```

# QQ Plot

```python
# Generate distribution
rand_vec = np.random.uniform(0,1,10000)

#Plot histogram
plt.hist(rand_vec)
plt.show()

#QQ Plot of distribution
stats.probplot(rand_vec, dist="norm", plot=pylab)
pylab.show()
stats.probplot(rand_vec, dist="uniform", plot=pylab)
pylab.show()
```

# Sampling distribution

* Question: Population **must** have a normal distribution?

```python
# Generate normal distribution
pop_dim = 1000000
population = np.random.normal(0,10,pop_dim) #(mean,standard deviation, dimension)

#Plot histogram
plt.hist(population, bins = 30)
plt.show()

# Generating samples
n_samples = 10
n_observation = 50
samples_data = np.zeros((n_samples, n_observation))

for i in range(n_samples):
    samples_data[i] = rnd.sample(list(population), n_observation)
samples_data = samples_data.T

c = [str(i) for i in range(n_samples)]
c = ['Sample ' + i for i in c]

df_samples = pd.DataFrame(samples_data, columns= c)
df_samples


# Compute dataset mean
print("Dataset mean", population.mean())

# Compute samples means
samples_means = df_samples.mean()
# print(samples_means)
plt.hist(samples_means, bins = 30)

print("Mean of sample mean: ", samples_means.mean())
```

As we can see, for *n_samples* sufficiently high, the mean of samples is the same as the mean of population (unless an error).

Note: the standard deviation of sample's means is the ***standard error***.
