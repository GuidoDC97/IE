```python
import numpy as np
import math as mt
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
# %matplotlib notebook
# from IPython.core.interactiveshell import InteractiveShell
```

# Exercise 1

```python
t_arr = np.arange(0,5,0.01)
lam = 0.5
Rmax = [y := (1-(1-mt.exp(-3*lam*t))**5) for t in t_arr]
Rsys = [y := (6*mt.exp(-3*lam*t) - 7*mt.exp(-4*lam*t) + 2*mt.exp(-5*lam*t) ) for t in t_arr]

plt.plot(t_arr, Rmax, label="$R_{sys_{up}}$")
plt.plot(t_arr, Rsys, label="$R_{sys}$")
plt.xlabel("Time")
plt.ylabel("Reliability")
plt.legend()
plt.savefig("Esercizio1_Rsysup_Rsys.svg", format = "svg")
```

# Exercise 2

```python
# System Reliability as function of r
r_arr = np.arange(0,1,0.01)

Rsys1 = [y := (1-(1-r**4)**2) for r in r_arr]
Rsys2 = [y := (1-(1-r)**2)**4 for r in r_arr]

plt.plot(r_arr, Rsys1, label="$R_{sys1}$")
plt.plot(r_arr, Rsys2 , label="$R_{sys2}$")
plt.xlabel("r")
plt.ylabel("Reliability")
plt.legend()
plt.savefig("Esercizio2_Rsys1_Rsys2_r.svg", format = "svg")
```

```python
# System Reliability as function of t
t_arr = np.arange(0,5,0.01)
lam = 0.5
Rsys1_t = [y := (1-(1-mt.exp(-lam*t)**4)**2) for t in t_arr]
Rsys2_t = [y := ((1-(1-mt.exp(-lam*t))**2)**4) for t in t_arr]

plt.plot(t_arr, Rsys1_t)
plt.plot(t_arr, Rsys2_t)
```

```python
# System Reliability
t_arr = np.arange(0,300,0.01)
lam = 0.005
Rsys = [y := (mt.exp(-8*lam*t) + 8*mt.exp(-7*lam*t)*(1-mt.exp(-lam*t)) + 20*mt.exp(-6*lam*t)*(1-mt.exp(-lam*t))**2 
              + 16*mt.exp(-5*lam*t)*(1-mt.exp(-lam*t))**3 + 2*mt.exp(-4*lam*t)*(1-mt.exp(-lam*t))**4) for t in t_arr]
Rsys48 = (mt.exp(-8*lam*48) + 8*mt.exp(-7*lam*48)*(1-mt.exp(-lam*48)) + 20*mt.exp(-6*lam*48)*(1-mt.exp(-lam*48))**2 
              + 16*mt.exp(-5*lam*48)*(1-mt.exp(-lam*48))**3 + 2*mt.exp(-4*lam*48)*(1-mt.exp(-lam*48))**4)

plt.plot(t_arr, Rsys, label="$R_{sys}(t)$")
plt.plot(48,Rsys48,"ro", label = "$R_{sys}(48)$")
plt.xlabel("Time")
plt.ylabel("Reliability")
plt.legend()

plt.savefig("Esercizio3_Rsys.svg", format = "svg")
```
